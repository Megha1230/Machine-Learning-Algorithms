```python
data['buying'],_=pd.factorize(data['buying'])
data['maint'],_=pd.factorize(data['maint'])
data['doors'],_=pd.factorize(data['doors'])
data['persons'],_=pd.factorize(data['persons'])
data['lug_boot'],_=pd.factorize(data['lug_boot'])
data['safety'],_=pd.factorize(data['safety'])
```

```python
data.head()
```

```python
x=data.iloc[:,0:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x.columns
```

```python
from sklearn import tree,metrics
dtree=tree.DecisionTreeClassifier(criterion='gini')
dtree.fit(x_train,y_train)
```

```python
y_pred=dtree.predict(x_train)
```

```python
acc=metrics.accuracy_score(y_train,y_pred)
print('Train Accuracy: {:.2f}'.format(acc))
```

```python
from sklearn import tree,metrics
dtree2=tree.DecisionTreeClassifier(criterion='gini')
dtree2.fit(x_test,y_test)
```

```python
y_test_pred=dtree2.predict(x_test)
acc1=metrics.accuracy_score(y_test,y_test_pred)
print('Test Accuracy: {:.2f}'.format(acc1))
```
