```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import sklearn
```

```python
test=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv")
train=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv")
```

```python
train.info()
test.info()
```

```python
train.set_index("PassengerId",inplace=True)
test.set_index("PassengerId",inplace=True)
```

```python
train
```

```python
y_train=train['Survived']
```

```python
train.drop(labels='Survived',axis=1,inplace=True)
```

```python
train.shape
```

```python
train_test=train.append(test)
```

```python
col_to_drop={'Name','Age','SibSp','Ticket','Cabin','Parch','Embarked'}
train_test.drop(['Name','Age','SibSp','Ticket','Cabin','Parch','Embarked'],axis=1,inplace=True)
```

```python
train_test
```

```python
train_test_dummies=pd.get_dummies(train_test,columns={'Sex'})
```

```python
train_test_dummies.shape
```

```python
train_test_dummies.isnull().sum()
```

```python
train_test_dummies.fillna(value=0,inplace=True)
```

```python
x_train=train_test_dummies.values[0:891]
x_test=train_test_dummies.values[891:]
```

```python
from sklearn.preprocessing import MinMaxScaler
sca=MinMaxScaler()
x_train_scale=sca.fit_transform(x_train)
x_test_scale=sca.transform(x_test)
```

```python
from sklearn.model_selection import train_test_split
x_train_sub,x_test_sub,y_train_sub,y_test_sub=train_test_split(x_train_scale,y_train,random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
learning_rates = [0.05,0.1,0.25,0.5,0.75,1]
for learning_rate in learning_rates:
  gb = GradientBoostingClassifier(n_estimators=20,learning_rate=learning_rate,max_features=2,max_depth=2,random_state=0)
  gb.fit(x_train_sub,y_train_sub)
  print("Learning rate:",learning_rate)
  print("Accuracy score(training) :{0:.3f}".format(gb.score(x_train_sub,y_train_sub)))
  print("Accuracy score (validation) : {0:.3f}".format(gb.score(x_test_sub,y_test_sub)))
  print()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
x_train_sub,x_test_sub,y_train_sub,y_test_sub=train_test_split(x_train,y_train,random_state=42)
base_model = DecisionTreeClassifier(max_depth=1)
adaboost_model = AdaBoostClassifier(base_model, n_estimators=50, random_state=42)
adaboost_model.fit(x_train, y_train)
y_pred = adaboost_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
